### **SOAL 1 - Load & Eksplorasi Data**

**Import Library**

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

**Load Dataset (Soal 1a)**

```python
df = pd.read_csv('anemia.csv')  # pastikan file sudah diupload ke Colab

df.head()
```

**Eksplorasi Data (Soal 1b)**

```python
print("Jumlah data:", df.shape[0])
print("Jumlah fitur:", df.shape[1] - 1)

print("\nNama fitur:")
print(df.columns[:-1].tolist())

print("\nDistribusi kelas target:")
print(df['Result'].value_counts())
```

**Penjelasan Potensi Masalah Data (Soal 1c)**

```python
print("""
Potensi Permasalahan Data:
1. Skala fitur berbeda (Hemoglobin, MCH, MCV, dll).
2. Algoritma KNN sensitif terhadap perbedaan skala fitur.
3. Perlu dilakukan feature scaling sebelum training.
""")
```

**Pisahkan Fitur & Target**

```python
X = df.drop(columns='Result')
y = df['Result']
```

### **SOAL 2 - Preprocessing**

**Split Data (Soal 2a)**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

**Feature Scaling (Soal 2b)**

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Penjelasan Scaling (Soal 2c)**

```python
print("""
Feature scaling penting karena:
1. KNN menghitung jarak antar data.
2. Fitur dengan nilai besar bisa mendominasi jarak.
3. Scaling membuat semua fitur memiliki pengaruh seimbang.
""")
```

### **SOAL 3 - KNN**

**Implementasi KNN (Soal 3a)**

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

**Latih Model & Prediksi (Soal 3b)**

```python
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
```

**Evaluasi Model (Soal 3c)**

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-Score :", f1)
```
