### **SOAL 1 - Load & Eksplorasi Data**

**Import Library**

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

**Load Dataset (Soal 1a)**

```python
data = load_breast_cancer()

X = data.data
y = data.target
```

**Eksplorasi Data (Soal 1b)**

```python
print("Jumlah data:", X.shape[0])
print("Jumlah fitur:", X.shape[1])
print("\nNama fitur:")
print(data.feature_names)

print("\nDistribusi kelas target:")
unique, counts = np.unique(y, return_counts=True)
for u, c in zip(unique, counts):
    print(f"Kelas {u}: {c} data")
```

**Penjelasan Potensi Masalah Data (Soal 1c)**

```python
print("""
Potensi Permasalahan Data:
1. Skala fitur berbeda-beda sehingga perlu dilakukan scaling.
2. Dataset relatif seimbang, namun tetap perlu dicek saat modeling.
3. Algoritma berbasis jarak seperti KNN sangat sensitif terhadap skala fitur.
""")
```

### **SOAL 2 - Preprocessing**

**Split Data (Soal 2a)**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

**Feature Scaling (Soal 2b)**

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Penjelasan Scaling (Soal 2c)**

```python
print("""
Feature scaling penting karena:
1. Algoritma K-Nearest Neighbor (K-NN) menggunakan jarak Euclidean untuk menentukan tetangga terdekat.
2. Fitur dengan nilai yang lebih besar dapat mendominasi perhitungan jarak jika tidak dilakukan scaling.
3. Scaling membuat seluruh fitur berada pada skala yang sebanding sehingga setiap fitur memiliki pengaruh yang adil dalam proses klasifikasi
""")
```

### **SOAL 3 - KNN**

**Implementasi KNN (Soal 3a)**

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

**Latih Model & Prediksi (Soal 3b)**

```python
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
```

**Evaluasi Model (Soal 3c)**

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-Score :", f1)
```
